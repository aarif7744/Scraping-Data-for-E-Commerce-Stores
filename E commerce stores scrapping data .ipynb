## Function to generate all  E commercw stores data

```python
def generate_all_data(num_rows):
    data={
        "E Commerce Stores Name": [],
        "Phone Number" : [],
        "Email Address":[],
        "Location": [],
        "Owner Name":[],
        "No. of Services":[]
   }
    for i in range(num_rows):
        e_commerce_stores_name = random.choice(e_commerce_stores_names)
        phone_number = f"+91-98765412{i%100:02d}"
        email_address = f"contact{(i+1)%100}@ecommerce{i%50}.com"
        location =random.choice(locations)
        owner_name=random.choice(owner_names)
        no_of_services = random.randint(i,50)
        
        data["E Commerce Stores Name"].append(e_commerce_stores_name)
        data["Phone Number"].append(phone_number)
        data["Email Address"].append( email_address)
        data["Location"].append( location)
        data["Owner Name"].append(owner_name)
        data["No. of Services"].append(no_of_services)
        
    return pd.DataFrame(data)
```

## Generate 50 rows of all E commerce stores data

```python
df =generate_all_data(50)
```

## Converted the E commers stores data into csv file

```python
df.to_csv("E commerce Stores.csv", index=False)
```

## Scatter plot

```python
plt.figure(figsize=(8,6))
for location in df['Location'].unique():
    subset = df[df['Location'] ==location]
    plt.scatter(subset['Location'],subset['No. of Services'],label=location)
plt.title('Number of Services Available by Location')
plt.xlabel("Location")
plt.ylabel('No.of Services Available')
plt.legend()
plt.show()
plt.grid()
```

## Bar plot

```python
plt.figure(figsize=(8,6))
sns.countplot(data=df ,x='Location')
plt.title('No.of E Commerce stores per Location')
plt.xlabel("Location")
plt.ylabel("No.of E Commerce store ")
plt.show()
```
